```python
import pandas as pd
```

```python
data = pd.read_csv('/kaggle/input/google-play-store-apps/googleplaystore.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#histogram for app ratings
plt.figure(figsize=(10,6))
sns.histplot(data,x='Rating',bins=30)
plt.xlabel('Rating')
plt.ylabel('count')
plt.title('app ratings')
plt.show()
```

**which app category has highest no of apps?**

```python
plt.figure(figsize=(10,6))
sns.countplot(data,y='Category',order = data['Category'].value_counts().index)
plt.xlabel('count')
plt.ylabel('Category')
plt.title('app categories')
plt.show()
```
